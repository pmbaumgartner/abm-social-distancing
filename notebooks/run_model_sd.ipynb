```python
model = SocialDistanceModel(10, 800, 400)

for _ in range(100):
    model.step()
```

```python
def unpack_position(pos):
    return pd.Series({"x" : pos[0], "y" : pos[1]})
```

```python
agent_positions = model.datacollector.get_agent_vars_dataframe()

position_split_df = agent_positions['pos'].apply(unpack_position).reset_index()
```

```python
position_split_df.head()
```

```python
alt.Chart(position_split_df).mark_point(opacity=0.3, size=50).encode(
    x=alt.X("x"),
    y=alt.Y("y"),
    tooltip=["Step", "AgentID"],
    color="Step",
    shape="AgentID:N",
)
```

```python
fig = px.scatter(position_split_df, x="x", y="y", size=[10] * len(position_split_df), size_max=10, animation_frame="Step", animation_group="AgentID", range_x=[0,800], range_y=[400,0])
fig.show()
fig.write_html("animation.html")
```
